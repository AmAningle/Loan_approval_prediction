### Encoding of caterogical features 

```python
from sklearn import preprocessing
import warnings
warnings.filterwarnings("ignore")  # Reset warnings to default behavior
enc=preprocessing.LabelEncoder()
train['Married']=enc.fit_transform(train['Married'])
train['Gender']=enc.fit_transform(train['Gender'])
train['Self_Employed']=enc.fit_transform(train['Self_Employed'])
train['Property_Area']=enc.fit_transform(train['Property_Area'])
train['Education']=enc.fit_transform(train['Education'])
```

```python
train.replace(to_replace='3+', value=4, inplace=True)            # replacing the value of 3+ to 4
```

```python

train.drop('Loan_ID',inplace=True,axis=1)

train
```

```python
y=train['Loan_Status']
x=train.iloc[:,0:11]
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
models=[]                                 # empty list created

models.append(("Logistic Regression",LogisticRegression()))                     # append various models
models.append(("Decision Tree",DecisionTreeClassifier()))
models.append(("Random Forest",RandomForestClassifier()))
models.append(("Support Vector Classifier",SVC()))
models.append(("K- Nearest Neighbour",KNeighborsClassifier()))
```

```python
models
```

```python
from sklearn.model_selection import KFold 
from sklearn.model_selection import cross_val_score
names=[]
result=[]
```

```python
from sklearn.model_selection import KFold, cross_val_score

# Assuming you have defined your models and x_train, y_train

for name, model in models:
    kfold = KFold(n_splits=10, random_state=0, shuffle=True)
    cv_result = cross_val_score(model, x_train, y_train, cv=kfold, scoring='accuracy')             
    result.append(cv_result)
    names.append(name)
    
    # Fit the model to the entire training set
    model.fit(x_train, y_train)
    
    # Make predictions on the test set
    y_pred = model.predict(x_test)
    
    # Calculate accuracy between predictions and y_test
    accuracy = accuracy_score(y_test, y_pred)
    
    
 
    print("%s - Cross-validated Accuracy: %f" % (name, cv_result.mean()))
    print("%s - Test Set Accuracy: %f" % (name, accuracy))
    print("=" * 30)  # Separating different models for better readability
```
